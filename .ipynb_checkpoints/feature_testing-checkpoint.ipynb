## 1. Probability of occurrence for key card

### 1.1. Description:

In a deck of `N` cards, there are `K` specific cards referred to as 'target cards.' From this deck, `D` cards are randomly drawn. Calculate the probability that at least one of the `D` drawn cards is a target card.

#### 1.1.1. Purpose: 

Help players understand how likely they are to draw essential cards early in the game.

#### 1.1.2. Customization: 
Allow players to define which cards are considered 'key' to their strategy.

### 1.2. Mathematical solution:

$$P_{\#key\not=0}(N,K,D)=1-\frac{N-K}{N}\frac{N-K-1}{N-1}\cdots\frac{N-K-D+1}{N-D+1}.$$

### 1.3. Codes:

Construct a tensor to represent the multidimensional table of solutions for the probability of occurrence for key card problem. The dimensions of the tensor correspond to the possible values of the input settings N, K, and D. For example, given input parameters `N=20, K=7, D=9`, the resulting tensor should have dimensions `21x8x10`, encapsulating all possible combinations of input values ranging from `(0,0,0)` to `(N,K,D)`. Each entry in the tensor, such as the one at index `15,3,4`, represents the solution to the probability of not drawing a target card for the corresponding input values, in this case `N=15, K=3, D=4`.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Calculate the probability directly
def calculate_probability_direct(N, K, D):
    probability = 1.0  # Initialize probability to 1
    for m in range(D):
        probability *= (N - K - m) / (N - m)
    return probability

# Returns a tensor giving all solutions of the input from (0,0,0) to (N,K,D)
# DP Method is applied
def calculate_probability(N, K, D):
    P = np.zeros((N + 1, K + 1, D + 1))
    P[:, :, 0] = 1  # Base case initialization
    # print(P)

    for n in range(1, N + 1):
        # print('n=',n)
        for k in range(1, n): # k<=n-1
            for d in range(1, n - k + 1):
                # print('n=',n,', k=',k,', d=',d,'\n')
                P[n, k, d] = (n - k - d + 1) / (n - d + 1) * P[n, k, d - 1]
    return 1-P

# Returns a tensor giving all solutions of the input from (0,0,0) to (N,K,D)
def prob_truth(N,K,D):
    P = np.zeros((N + 1, K + 1, D + 1))
    P[:, :, 0] = 1  # Base case initialization
    for n in range(1, N + 1):
        # print('n=',n)
        for k in range(1, n): # k<=n-1
            for d in range(1, n - k + 1):
                # print('n=',n,', k=',k,', d=',d,'\n')
                P[n, k, d] = calculate_probability_direct(n, k, d)
    return 1-P

# Example usage
N = 20  # Total cards
K = 20  # Cards of interest
D = 20   # Number of draws
probability_tensor = calculate_probability(N, K, D)
probability_truth = prob_truth(N,K,D)
# print(probability_tensor)
# print(np.array_equal(probability_truth,probability_tensor))
```

Search for an answer with setting `N=20,K=20,D=20`

```python
print(probability_tensor[20,7,1])
print(probability_truth[20,7,1])
print(calculate_probability_direct(20,7,1))
```

```python
# Creating a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Preparing data for visualization
x, y, z = np.indices((N + 1, K + 1, D + 1))

# Plotting
scatter = ax.scatter(x.flatten(), y.flatten(), z.flatten(), c=probability_truth.flatten(), cmap='YlOrBr', alpha=0.9)
ax.set_xlabel('Total Cards (N)')
ax.set_ylabel('Cards of Interest (K)')
ax.set_zlabel('Draws (D)')
plt.colorbar(scatter, ax=ax, label='Probability')

plt.title('Probability of Not Drawing a Card of Interest')
plt.show()
```

## 2. Probability of occurrence for a group of key cards 

Assess the chance of drawing specific combinations of key cards by a certain turn.

### 2.1. Description:

In a deck of `N` cards, there are `K` specific cards referred to as 'target cards.' From this deck, `D` cards are randomly drawn. Calculate the probability that `G` of the `D` drawn cards are target cards.

#### 2.1.1. Purpose: 
Enable players to evaluate the feasibility of drawing combinations that could lead to a win.

#### 2.1.2. Complexity: 
Consider different combinations and their impact on game strategy.

### 2.2. Mathematical solution:
Hypergeometric distribution:

$$P(N,K,D,G)=\frac{\frac{K!}{G!(K-G)!}\frac{(N-K)!}{(D-G)!(N-K-D+G)!}}{\frac{N!}{D!(N-D)!}}$$

### 2.3. Codes:

Source: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.hypergeom.html

```python
import numpy as np
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
```

```python
[N, K, D] = [20, 7, 12]
rv = hypergeom(N, K, D)
g = np.arange(0, K+1)
pmf_dogs = rv.pmf(g)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(g, pmf_dogs, 'bo')
ax.vlines(g, 0, pmf_dogs, lw=2)
ax.set_xlabel('# of dogs in our group of chosen animals')
ax.set_ylabel('hypergeom PMF')
plt.show()
```

## 3. Probability of occurrence for functioning structure of tags on cards  


### 3.1. Description:

Consider a deck composed of `N` cards, each potentially labeled with one or more of `T` distinct tags. From this deck, we randomly draw `D` cards, ensuring that every card has an equal chance of being selected. A key aspect to consider is that cards may carry multiple tags, and a single tag can be present on several cards. Our focus centers on particular tag combinations or structures on these cards. The reason for this focus is strategic: certain tag groups, when combined, create synergies that lead to favorable outcomes in gameplay. Therefore, the objective here is to calculate the probability that the `D` cards drawn will, together, form at least one of these strategically advantageous tag structures.

#### 3.1.1. Purpose: 

Enable deck builders to quantify how likely it is that their decks, when put into play, will exhibit some effective tag combinations.

By determining the probability that the `D` cards drawn will include at least one group of tags that works perfectly together on the board, deck builders can strategically optimize their decks for maximum efficiency and effectiveness according to their design and gameplay strategy.

#### 3.1.2. Customization: 
Allow players to tag cards and define which tag combinations are considered effective and desired to their strategy.

### 3.2. Mathematical solution:

Addressing this problem involves employing the multivariate hypergeometric distribution, which is particularly suited for situations where objects are drawn from a finite population without replacement, as is the case with our deck of cards. The key here is to recognize that each 'desired tag structure' can be treated as a distinct outcome within this distribution.

The probability mass function (PMF) of the multivariate hypergeometric distribution gives us the probability of drawing a specific combination of tagged cards. For our purposes, each combination corresponds to one of the desired tag structures.

The PMF for the multivariate hypergeometric distribution is given as:

$$ P(X_1 = x_1, X_2 = x_2, ..., X_k = x_k|N,D) = \frac{{\prod_{i=1}^{k} \binom{n_i}{x_i}}}{\binom{N}{D}} $$

In this formula, the term $ k $ equals $ 2^T-1 $, signifying the total count of all possible non-empty subsets of tags, given $ T $ distinct tags. $ \binom{n_i}{x_i} $ denotes the binomial coefficient. Here, $ n_i $ represents the total number of cards in the deck that belongs to the $i$-th subset of tags. On the other hand, $ x_i $ indicates the number of cards belongint to the $i$-th subset of tags among the cards drawn. In this context, $ N $ is the total number of cards in the entire deck, while $ D $ refers to the number of cards that are drawn. Note that 

$$\sum_{i=1}^k x_i = D\Leftrightarrow x_k=D-\sum_{i=1}^{k-1} x_i.$$

To find the overall probability of drawing at least one of the desired tag structures, we sum the probabilities of each specific tag structure as calculated using the above PMF:

$$P_S(N,D)=\sum_{(x_1,x_2,\cdots,x_k)\in S}P(x_1, x_2, \cdots, x_k|N,D)$$

where $S$ denotes the set consisting of all desired tag combinations in terms of the distribution of correspoding to subsets of tags.


This approach enables us to determine our target probability, reflecting how likely it is for a randomly drawn set of cards to form a strategically advantageous arrangement.

### 3.3. Codes:

```python
from scipy.stats import multivariate_hypergeom

# Define the population sizes for each category
population_sizes = [10, 15, 5]  # For example, 10 red, 15 blue, 5 green

# Define the number of draws
draws = 6

# Define the outcome you're interested in
desired_outcome = [2, 3, 1]  # For example, 2 red, 3 blue, 1 green

# Calculate the probability
probability = multivariate_hypergeom.pmf(desired_outcome, population_sizes, draws)

print(f"The probability of drawing {desired_outcome} from {population_sizes} in {draws} draws is: {probability}")
```
